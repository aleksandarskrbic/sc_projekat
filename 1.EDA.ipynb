```python
def show_examples(path):
    fig = plt.figure(figsize=(20, 10))
    cols = 3
    rows = 2
    cnt = 0
    
    for img_id in os.listdir(path):
        if cnt == 6:
            break;
        else:
            cnt += 1
            for i in range(1, 7):
                img = cv2.imread(f'{path}/{img_id}')
                fig.add_subplot(2, 3, i)
                plt.imshow(img)
    plt.show()
```
